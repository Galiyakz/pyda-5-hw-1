Задание 1 <br>
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:

оценка 2 и меньше - низкий рейтинг<br>
оценка 4 и меньше - средний рейтинг<br>
оценка 4.5 и 5 - высокий рейтинг <br>
Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
ratings = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
movies.head()
```

```python
ratings.head()
```

```python
mov_rat = pd.merge(movies, ratings, on=['movieId', 'movieId'], how='inner')
mov_rat.head()
```

```python
def classification(rating):
    if rating <= 2:
        return 'низкий рейтинг'
    if rating <= 4:
        return 'средний рейтинг'
    elif rating == 4.5 or rating == 5.0:
        return 'высокий рейтинг'
```

```python
mov_rat['class'] = mov_rat['rating'].apply(classification)
mov_rat.head()
```

```python
mov_rat['first_genre'] = mov_rat['genres'].apply(lambda row: row.split('|'))
```

```python
mov_rat.head()
```

Задание 2 <br>
Используем файл keywords.csv.
Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.
Правила распределения по регионам Центр, Северо-Запад и Дальний Восток: <br>
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
Результат классификации запишите в отдельный столбец region.

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
kw = pd.read_csv('ml-latest-small/keywords.csv')
kw.head()
```

```python
def defRegion(row): # вк
    for key, value in geo_data.items(): # ''Северо-Запад': ['петербург', 'псков', 'мурманск'],
        if row['keyword'] in value: # петербург in ['петербург', 'псков', 'мурманск'],
            return key # Северо-Запад
        
    return 'undefined'
```

```python
kw['region'] = kw.apply(defRegion, axis=1)
```
